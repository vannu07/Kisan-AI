# 02. Model Training / Prompt Engineering

## Using Google Gemini Pro Vision

For KisanAI, we leverage the pre-trained multimodal capabilities of Gemini Pro Vision.
Instead of traditional gradient-based training, we perform **Prompt Engineering** to optimize the model's performance on agricultural tasks.

```python
import google.generativeai as genai
import os
from dotenv import load_dotenv
from PIL import Image

load_dotenv()
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

# Initialize Model
model = genai.GenerativeModel('gemini-pro-vision')

# Define System Prompt (The "Training" equivalent)
system_prompt = """
You are an expert plant pathologist.
Analyze the image and provide:
1. Disease Name
2. Confidence Level
3. Treatment Plan
"""
```
