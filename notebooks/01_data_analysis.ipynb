## KisanAI Live - Agriculture Dataset Analysis

In this notebook, we explore the agriculture dataset used for crop disease detection.
Although we are using Gemini Vision (Zero-shot/Few-shot), understanding the data distribution is crucial for prompt engineering.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from pathlib import Path

# Define paths
RAW_DATA_PATH = Path("../artifacts/raw_data")

# Mock dataset loading (since we are using real-time uploads in the app)
# In a real training scenario, this would load a CSV of labeled images
data = {
    'crop': ['Tomato', 'Potato', 'Corn', 'Rice', 'Wheat'],
    'disease': ['Early Blight', 'Late Blight', 'Healthy', 'Leaf Blast', 'Rust'],
    'count': [150, 120, 200, 180, 160]
}

df = pd.DataFrame(data)
print(df.head())
```

```python
# Visualize Class Distribution
plt.figure(figsize=(10, 6))
plt.bar(df['disease'], df['count'], color='green')
plt.title('Distribution of Crop Diseases in Training Set')
plt.xlabel('Disease Type')
plt.ylabel('Image Count')
plt.show()
```
